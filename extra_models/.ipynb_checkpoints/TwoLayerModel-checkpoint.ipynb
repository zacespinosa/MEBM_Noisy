```python
import numpy as np
import scipy.io as io
import matplotlib.pyplot as plt
```

```python
#variables to modify:
gamma = 0.7 #ocean heat uptake efficiency parameter (W/m2/K), strength of coupling between layers
lambda0 = -1.2 #radiative feedback parameter (W/m2/K)
efficacy = 1 #ocean heat uptake efficacy (unitless)
h1 = 70 #upper ocean depth
h2 = 1100 #deep ocean depth
R = 4 #forcing for CO2 doubling

num_years = 1000 #length of run in years

#constants:
rho=1025; #density of sea water kg/m3
cw=3985; #specific heat of sea water J/Kg/C
secondsperyear = 30.25*24*60*60*12

c1 = rho * cw * h1
c2 = rho * cw * h2

years = np.arange(num_years)
T1 = np.zeros(num_years)
T2 = np.zeros(num_years)

for i in range(num_years - 1):
    T1[i+1] = T1[i] + ((lambda0 * T1[i] + R + gamma*efficacy*(T2[i] - T1[i]))/c1 * secondsperyear)
    T2[i+1] = T2[i] + ((gamma * (T1[i] - T2[i]))/ c2 * secondsperyear)
    
toa_imbalance = R + lambda0*T1 - (efficacy-1)*gamma*(T1-T2) #top of atmosphere radiative imbalance

legend = ['Surface Layer', 'Deep Ocean Layer']
fig, ax = plt.subplots(1,1, figsize = (16,8))
ax.plot(years, T1)
ax.plot(years, T2)
plt.xlabel('Years', size = 'xx-large')
plt.ylabel('Temperature Anomaly (K)', size = 'xx-large')
ax.legend(legend)

fig, ax = plt.subplots(1,1, figsize = (16,8))
ax.plot(T1, toa_imbalance)
plt.xlabel('TOA imbalance (W/m2/K)', size = 'xx-large')
plt.xlabel('Temperature Anomaly (K)', size = 'xx-large')
ax.legend(legend)
```
